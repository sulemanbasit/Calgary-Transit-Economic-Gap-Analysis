```python
import requests
from bs4 import BeautifulSoup
import os
# import wget
import re
```

```python
folder_name = "Community Profiles"
if not os.path.exists(folder_name):
    os.makedirs(folder_name, exist_ok=True)
else:
    print("{} already exists".format(folder_name))
```

```python
base_url = "https://www.calgary.ca/communities/profiles.html"
# Send a GET request to the URL
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')
# print(response.status_code)
cmnt = soup.select('div.text.parbase a')
print(cmnt)
```

```python
clean_link = []
for link in cmnt:
    if re.search(r'/communities/profiles/(.+).html', str(link)):
        link_string = re.sub(r'<a href="(.*)">(.*)</a>', r'\1',str(link))
        if "wards" not in link_string:
            if "title" in link_string:
                index = link_string.find("title")
                clean_link.append(str(link_string[:(index-2)]))
            else:
                clean_link.append(str(link_string))

for i in clean_link:
    print(i)
```

```python
# os.chdir(os.getcwd()+"/Community Profiles")
com_link_list =[]
for lnk in clean_link:
    sub_url = base_url[:22] + lnk
    # print(sub_url)
    new_response = requests.get(sub_url)
    soup = BeautifulSoup(new_response.text, 'html.parser')
    com_link = soup.findAll('div', 'text parbase')
    com_link_list.append(str(com_link))
    # print(com_link)
    match = re.search(r'href="(.*?)"', str(com_link))
    if match:
        sub_link = match.group(1)
        sub_pdf_url = base_url[:22] + sub_link
        # print(sub_link[83:])
        output_name = sub_link[83:]
        last_response = requests.get(sub_pdf_url)
        if last_response.status_code == 200:
            with open(output_name, 'wb') as file:
                file.write(last_response.content)
                print(f"File '{output_name}' downloaded successfully.")
        else:
            print(f"Failed to download file. Status code: {last_response.status_code}")
```
